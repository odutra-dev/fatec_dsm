# Trabalhando com Tokens em PLN

## 1. Procurar e fazer download de um texto qualquer na internet e salvar como TXT.

```python
import requests
```

```python
# 1. Baixar um texto da internet (ex: Projeto Gutenberg)
url = 'https://www.gutenberg.org/cache/epub/1513/pg1513.txt'  # Romeu e Julieta
response = requests.get(url)
texto = response.text

# 2. Salvar como arquivo .txt
with open('Romeu e Julieta.txt', 'w', encoding='utf-8') as f:
    f.write(texto)
```

## 2. Faça a tokenização de sentenças.

```python
import nltk
```

```python


# Baixar recursos do NLTK se ainda não estiverem disponíveis
nltk.download()
```

```python
# 3. Tokenização de sentenças
sentencas = nltk.sent_tokenize(texto)

print(sentencas)
```

## 3. Faça a tokenização de palavras.

```python
# 4. Tokenização de palavras
palavras = nltk.word_tokenize(texto)

print(palavras)
```

```python
# Limpar pontuações e transformar em minúsculas
palavras_limpa = [palavra.lower() for palavra in palavras if palavra.isalpha()]

print(palavras_limpa)
```

## 4. Mostre a quantidade de sentenças e palavras encontradas.

```python
# 5. Mostrar quantidades
print(f"Número de sentenças: {len(sentencas)}")
print(f"Número de palavras: {len(palavras)}")
print(f"Número de palavras limpas: {len(palavras_limpa)}")
```

## 5. Mostre a frequência das palavras em formato de gráfico de barras (matplotlib).

```python
from collections import Counter
import matplotlib.pyplot as plt
```

```python
# 6. Frequência das palavras
frequencia = Counter(palavras)
mais_comuns = frequencia.most_common(20)
```

```python
# 7. Frequência das palavras Limpas
frequencia_limpa = Counter(palavras_limpa)
mais_comuns_limpa = frequencia.most_common(20)
```

```python
# 8. Gráfico com matplotlib
palavras_grafico = [item[0] for item in mais_comuns]
contagens = [item[1] for item in mais_comuns]

plt.figure(figsize=(12, 6))
plt.bar(palavras_grafico, contagens, color='skyblue')
plt.title('Top 20 Palavras Mais Frequentes')
plt.xlabel('Palavras')
plt.ylabel('Frequência')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 9. Gráfico com matplotlib e palavras limpas
palavras_grafico_limpa = [item[0] for item in mais_comuns_limpa]
contagens_limpa = [item[1] for item in mais_comuns_limpa]

plt.figure(figsize=(12, 6))
plt.bar(palavras_grafico_limpa, contagens_limpa, color='skyblue')
plt.title('Top 20 Palavras Mais Frequentes')
plt.xlabel('Palavras')
plt.ylabel('Frequência')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
